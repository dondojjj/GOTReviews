```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
#have to know te number of seasons and te url of the show - this example is for game of thrones

reviews = []
for season in range(8):
# Make a request to the webpage
    #get season reviews
    print(f"retrieving season {str(season+1)} reviews")
    url = f"https://www.rottentomatoes.com/tv/game_of_thrones/s0{str(season+1)}/reviews?intcmp=rt-what-to-know_read-critics-reviews"
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    #get the pagination manager id to call the api
    pagination_manager = soup.find('pagination-manager')
    #print(pagination_manager['endpoint'])
    uri = pagination_manager['endpoint']
    nextPageUri=''
    #print(uri)
    nextPage = True
    while nextPage:
        url = f"https://www.rottentomatoes.com/{uri+nextPageUri}"
        response_reviews = requests.get(url)
        #print(response)
        content = response_reviews.json()
        nextPage = content['pageInfo']['hasNextPage']
        #print(nextPage)
        for review in content['reviews']:
            review_dict = {'season' :str(season+1),
                            'episode' :'0',
                            'EpisodeTitle': 'N/A',
                            'Rating': review['reviewState'],
                            'Content': review['quote'],
                            'Date': review['creationDate'],
                            'sentiment': review['scoreSentiment']
                                }
            reviews.append(review_dict)
        if nextPage:
            nextPageUri = f"?after={content['pageInfo']['endCursor']}&pagecount=20" 
    #get season episodes
    url = f"https://www.rottentomatoes.com/tv/game_of_thrones/s0{str(season+1)}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    episodes = soup.find_all('a',class_='unset title')

    #get episodes reviews
    for episode in episodes:
        url = f"https://www.rottentomatoes.com/tv/game_of_thrones/{episode['href']}/reviews?intcmp=rt-what-to-know_read-critics-reviews"
        print(f"retrieving {episode['href']} reviews")
        response = requests.get(url)

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        #gete the pagination manager id to call the api
        pagination_manager = soup.find('pagination-manager')
        #print(pagination_manager['endpoint'])
        uri = pagination_manager['endpoint']
        nextPageUri=''
        #print(uri)
        nextPage = True
        while nextPage:
            url = f"https://www.rottentomatoes.com/{uri+nextPageUri}"
            response_reviews = requests.get(url)
            #print(response)
            content = response_reviews.json()
            nextPage = content['pageInfo']['hasNextPage']
            #print(nextPage)
            for review in content['reviews']:
                review_dict = {'season' :str(season+1),
                                'episode' : episode.get_text().strip().split('.')[0],
                                'EpisodeTitle': episode.get_text().strip(),
                                'Rating': review['reviewState'],
                                'Content': review['quote'],
                                'Date': review['creationDate'],
                                'sentiment': review['scoreSentiment']
                                    }
                reviews.append(review_dict)
            if nextPage:
                nextPageUri = f"?after={content['pageInfo']['endCursor']}&pagecount=20" 
# Convert the list of reviews into a pandas dataframe
reviews_df = pd.DataFrame(reviews)
reviews_df.to_csv('GOTCriticsReviews.csv',index=False)
# Print the first 5 rows of the dataframe
print(reviews_df.head())
print(reviews_df.shape)
```
